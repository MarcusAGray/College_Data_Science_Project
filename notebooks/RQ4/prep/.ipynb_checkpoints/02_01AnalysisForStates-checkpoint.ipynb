### Try to calculate estimate

```python
diabetes_diagnosis_percentage_state_df.head()
```

```python
diabetes_diagnosis_percentage_state_df['Death Per Pop'] = states_final_df['Diabetes Death per Pop'].copy()
diabetes_diagnosis_percentage_state_df['Deaths'] = states_final_df['Diabetes Deaths'].copy()

diabetes_diagnosis_percentage_state_df
```

```python
#Have to clarify what I am doing here
diabetes_diagnosis_percentage_state_df['Death Per Pop/Diagnosis Percentage'] = diabetes_diagnosis_percentage_state_df['Death Per Pop'] / diabetes_diagnosis_percentage_state_df['Diagnosed Diabetes Percentage']
diabetes_diagnosis_percentage_state_df['State'] = diabetes_diagnosis_percentage_state_df['State'].str.upper()
diabetes_diagnosis_percentage_state_df.head()
```

```python
df3 = states_final_df[['State', 'Population']].copy()
df3.head()
```

```python
diabetes_diagnosis_percentage_state_df = diabetes_diagnosis_percentage_state_df.merge(df3, on=['State'])
diabetes_diagnosis_percentage_state_df.head()
```

```python
#find number of population diagnosed
diabetes_diagnosis_percentage_state_df['Number of Pop Diagnosed'] = (diabetes_diagnosis_percentage_state_df['Diagnosed Diabetes Percentage'] / 100) * diabetes_diagnosis_percentage_state_df['Population']
diabetes_diagnosis_percentage_state_df.head()
```

```python
#Calculate deaths per number of population diagnosed
diabetes_diagnosis_percentage_state_df['Deaths per # Diagnosed'] = diabetes_diagnosis_percentage_state_df['Deaths'] / diabetes_diagnosis_percentage_state_df['Number of Pop Diagnosed']
diabetes_diagnosis_percentage_state_df.head()
```

```python
#Can use the lower and upper quartile for estimates
diabetes_diagnosis_percentage_state_df['Deaths per # Diagnosed'].describe()
```

```python
diabetes_diagnosis_percentage_state_df['Deaths per # Diagnosed'].quantile(0.25)
```

```python
lower_q = diabetes_diagnosis_percentage_state_df['Deaths per # Diagnosed'].quantile(0.25)
%store lower_q
```

```python
mean = diabetes_diagnosis_percentage_state_df['Deaths per # Diagnosed'].mean()
%store mean
```

```python
upper_q = diabetes_diagnosis_percentage_state_df['Deaths per # Diagnosed'].quantile(0.75)
%store upper_q
```
