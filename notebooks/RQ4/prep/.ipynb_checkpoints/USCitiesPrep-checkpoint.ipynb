```python
#Here we iterate through the cities and sort them into a list of lists corresponding to the MSA areas they are in.
#We need to know the MSA areas because these contain the states the cities are located
city_list = []

temp_list = city_crime_rate_df['County/City'].dropna().tolist()

i = 0
l = []
while i < len(temp_list):
   
    if 'City' in temp_list[i]:
        l.append(temp_list[i])
        if 'City' not in temp_list[i + 1]:
            city_list.append(l)
            l = []
    i = i + 1
    
city_list
```

```python
city_crime_rate_df['State'] = city_crime_rate_df['State Code'].map(state_dict)
city_crime_rate_df
```

```python
city_crime_rate_df = city_crime_rate_df[city_crime_rate_df['County/City'].notnull()]
city_crime_rate_df
```

```python
city_crime_rate_df = city_crime_rate_df[['County/City', 'State','Population', 'Violent Crime', 'Murder and nonnegligent manslaughter']]

city_crime_rate_df
```

```python
city_crime_rate_df = city_crime_rate_df[city_crime_rate_df['County/City'].str.contains("Includes")==False]
city_crime_rate_df = city_crime_rate_df[city_crime_rate_df['County/City'].str.contains("Rate per")==False]
city_crime_rate_df = city_crime_rate_df[city_crime_rate_df['County/City'].str.contains("Estimated")==False]
city_crime_rate_df = city_crime_rate_df[city_crime_rate_df['County/City'].str.contains("Total area")==False]
city_crime_rate_df = city_crime_rate_df.reset_index(drop=True)
city_crime_rate_df
```

```python
temp_df = city_crime_rate_df['County/City'].str.split(' ', n = 2, expand = True)
city_crime_rate_df["County/City"] = temp_df[2] 
city_crime_rate_df
```

```python
city_crime_rate_df = city_crime_rate_df.rename(columns = {"County/City": "City"})
city_crime_rate_df
```
