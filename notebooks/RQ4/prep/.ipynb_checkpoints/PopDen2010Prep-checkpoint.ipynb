```python
import pandas as pd
import numpy as np
import re
```

```python
df = pd.read_excel("../../data/raw/POP01.xls", sheet_name="POP01C")
df
```

```python
df
```

Create dataset for State

```python
key_cols = ['Area_name','POP060210D']#column name for pop density of 2010
df = df[key_cols]
df.head()
```

```python
state_pop_den = df[~df['Area_name'].str.contains('UNITED STATES')]
state_pop_den = state_pop_den[~state_pop_den['Area_name'].str.contains(',')]
state_pop_den = state_pop_den.reset_index(drop=True)
state_pop_den
```

```python
state_pop_den = state_pop_den.rename(columns = {"POP060210D": "Pop Density"})
state_pop_den
```

```python
state_pop_den.to_csv("state_pop_den.csv", index=False)
```
